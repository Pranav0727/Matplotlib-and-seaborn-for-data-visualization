### Annotation

```python
batter= pd.read_csv('batter.csv')
```

```python
sample_df=batter.head(100).sample(25,random_state=5)
```

```python
sample_df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])

```

```python
# Horizontal and vertical line
```

```python
plt.figure(figsize=(15,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])

plt.axhline(130,color='r')
plt.axvline(32,color='g')
```

Subplot

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,6))

ax[0].scatter(batter['runs'],batter['strike_rate'],color='r')
ax[1].scatter(batter['avg'],batter['runs'],marker='*')

ax[0].set_title('scatter plot')
ax[0].set_xlabel('runs')
ax[0].set_ylabel('strike_rate')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))


ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

3D PLOT

```python
fig=plt.figure()

ax=plt.subplot(projection='3d')
ax.scatter(batter['runs'],batter['strike_rate'],batter['avg'],color='r')
```

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
x=[1,2,3,4]
y=[5,6,7,8]
z=[10,34,76,88]

ax.plot3D(x,y,z,color='r',marker='*')
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy=np.meshgrid(x,y)

z = xx**2 + yy**2
z.shape[0]


```

```python
fig =plt.figure()

ax=plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

Heatmap


```python
ipl=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

```python
ipl.sample(5)
```

```python
ipl_df=ipl[(ipl['ballnumber'].isin([1,2,3,4,5,6])) & (ipl['batsman_run']==6)]
```

```python
ipl_df
```

```python
grid =ipl_df.pivot_table(index='overs',values='batsman_run',columns='ballnumber',aggfunc='count')
```

```python
fig =plt.figure(figsize=(10,10))
plt.imshow(grid)
plt.yticks(ipl['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
