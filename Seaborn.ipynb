1.Relational Plot


1.Scatterplot 2.Lineplot

```python
tips= sns.load_dataset('tips')
tips.sample(5)
```

axis level function

```python
sns.scatterplot(data=tips ,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

Figure Level Function

```python
sns.relplot(data=tips ,x='total_bill',y='tip',hue='sex',style='smoker',size='size',kind='scatter')
```

Faceplot

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',row='day')
```

3.Matrixplot

1.Headmap 2.Clustermap

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
gap =px.data.gapminder()
gap.sample(5)
```

```python
temp_df=gap.pivot_table(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='winter')
```

```python

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
